```python
##Seaborn Visualization Examples by Farzam Salimi#######

import seaborn as sns
import matplotlib.pyplot as plt

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Create a bar plot of total bill by day
sns.barplot(x="day", y="total_bill", data=tips)

# Show the plot
plt.show()
```

Box plot showing the distribution of total bill by day:

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

Violin plot showing the distribution of total bill by day and split by sex:

```python
sns.violinplot(x="day", y="total_bill", hue="sex", split=True, data=tips)
plt.show()
```

Swarm plot showing the distribution of total bill by day and colored by sex:

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
plt.show()
```

Pair plot showing pairwise relationships between variables in the dataset:

```python
sns.pairplot(tips)
plt.show()
```

Line plot showing the relationship between total bill and tip:

```python
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.show()
```

Scatter plot showing the relationship between total bill and tip, colored by sex:

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.show()
```

Histogram showing the distribution of total bill:

```python
sns.histplot(x="total_bill", data=tips)
plt.show()
```

KDE plot showing the distribution of total bill:

```python
sns.kdeplot(x="total_bill", data=tips)
plt.show()
```

ECDF plot showing the empirical cumulative distribution of total bill:

```python
sns.ecdfplot(x="total_bill", data=tips)
plt.show()
```

Reg plot showing the linear regression between total bill and tip:

```python
sns.regplot(x="total_bill", y="tip", data=tips)
plt.show()
```

Resid plot showing the residuals of a linear regression between total bill and tip:

```python
sns.residplot(x="total_bill", y="tip", data=tips)
plt.show()
```

Strip plot showing the distribution of total bill by day:

```python
sns.stripplot(x="day", y="total_bill", data=tips)
plt.show()
```

Point plot showing the mean total bill by day and split by sex:

```python
sns.pointplot(x="day", y="total_bill", hue="sex", data=tips)
plt.show()
```

Count plot showing the count of observations by day:

```python
sns.countplot(x="day", data=tips)
plt.show()
```

```python
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd

# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Perform k-means clustering on the iris dataset
kmeans = KMeans(n_clusters=3)
kmeans.fit(iris_df)
iris_df['cluster'] = kmeans.labels_

# Create a pair plot colored by cluster
sns.pairplot(iris_df, hue='cluster')
```

Scatter plot showing the relationship between two variables in the iris dataset, colored by cluster:

```python
sns.scatterplot(x="sepal length (cm)", y="sepal width (cm)", hue="cluster", data=iris_df)
plt.show()
```

FacetGrid showing the relationship between two variables in the iris dataset, faceted by cluster:

```python
g = sns.FacetGrid(iris_df, col="cluster")
g.map(sns.scatterplot, "sepal length (cm)", "sepal width (cm)")
plt.show()
```

lmplot showing the linear regression between two variables in the iris dataset, colored by cluster:

```python
sns.lmplot(x="sepal length (cm)", y="sepal width (cm)", hue="cluster", data=iris_df)
plt.show()
```

Clustermap showing a clustered heatmap of the iris dataset:

```python
sns.clustermap(iris_df.drop('cluster', axis=1))
plt.show()
```
